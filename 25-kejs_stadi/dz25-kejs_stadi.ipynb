### Как критики относятся к спортивным играм?

Сформулируем нулевую и альтернативную гипотезы:

    H0: средняя удовлетворенность критиков спортивными играми = 7
    H1: средняя удовлетворенность критиков спортивными играми составляет менее 7
    
Поскольку в данном тесте цена ошибки не велика, зададим пороговый уровень статистической значимости alpha = 0.05.

Будем проводить одновыборочный тест, поскольку анализируем одну выборку и значение по выборке будет сравниваться с константой.

Будем проводить односторонний тест, нас интересует отклонение среднего значения удовлетворенности критиков только в одну сторону. Т.е., если среднее значение удовлетворенности критиков спортивными играми составляет менее 7, то нулевая гипотеза опровергается.

```python
from scipy import stats as st

# Подготовим данные
# Сформируем датафрейм с данными, где проставлена оценка критиков по спортивным играм.
df = games.loc[games.Genre == 'Sports'].loc[games.Critic_Score > 0] 

alpha = 0.05
result = st.ttest_1samp(df['Critic_Score'], 7)
print(result)

if (result.statistic < 0) & (result.pvalue / 2 < alpha):
    print('Отвергаем нулевую гипотезу, средняя удовлетворенность критиков спортивными играми меньше 7')
else:
    print('Не отвергаем нулевую нулевую гипотезу, критикам нравятся спортивные игры')
```

### Критикам нравятся больше игры на PC или на PS4?

Сформулируем нулевую и альтернативную гипотезы:

    H0: средняя удовлетворенность критиков играми на PC и PS4 одинаковая
    H1: средняя удовлетворенность критиков играми на PC и PS4 разная
    
Поскольку в данном тесте цена ошибки не велика, зададим пороговый уровень статистической значимости alpha = 0.05.

Будем проводить двухвыборочный двухсторонний тест, поскольку у нас есть две независимые выборки (если мы анализируем в общем игры на двух разных платформах, т.е. не анализируем одинаковые игры на двух разных платформах).

```python
# Подготовим данные
# Сформируем датафреймы с данными для PC и PS4, где проставлена оценка критиков по играм.
df_PC = games.query('Platform == "PC" and Critic_Score > 0')
df_PS4 = games.query('Platform == "PS4" and Critic_Score > 0')

alpha = 0.05
result = st.ttest_ind(df_PC['Critic_Score'], df_PS4['Critic_Score'], equal_var=False)
print(result)    
    
if (result.pvalue < alpha):
    print('Отвергаем нулевую гипотезу.')
    if (result.statistic < 0):
        print('Критикам больше нарвятся игры на PS4')
    elif (result.statistic > 0):
        print('Критикам больше нарвятся игры на PC')
else:
    print('Подтверждаем нулевую гипотезу. Критики одинаково относятся к играм на PC и PS4')
```

### Критикам больше нравятся стрелялки или стратегии?

Сформулируем нулевую и альтернативную гипотезы:

    H0: средняя удовлетворенность критиков по стрелялкам и стратегиям одинакова
    H1: средняя удовлетворенность критиков по стрелялкам и стратегиям разная

Поскольку в данном тесте цена ошибки не велика, зададим пороговый уровень статистической значимости alpha = 0.05.

Будем проводить двухвыборочный двухсторонний тест, поскольку у нас есть две независимые выборки.

```python
# Подготовим данные
# Сформируем датафреймы с данными для стрелслкам и стратегиям, где проставлена оценка критиков по играм.
df_Shooter = games.query('Genre == "Shooter" and Critic_Score > 0')
df_Strategy = games.query('Genre == "Strategy" and Critic_Score > 0')

alpha = 0.05
result = st.ttest_ind(df_Shooter['Critic_Score'], df_Strategy['Critic_Score'], equal_var=False)
print(result)    
    
if (result.pvalue < alpha):
    print('Отвергаем нулевую гипотезу.')
    if (result.statistic < 0):
        print('Критикам больше нарвятся стратегии')
    elif (result.statistic > 0):
        print('Критикам больше нарвятся стрелялки')
else:
    print('Подтверждаем нулевую гипотезу. Критики одинаково относятся к стрелялкам и стратегиям')
```

```python
# Создаем список сообщений из категории spam и список сообщений из категории ham
spam = list(df.loc[df.Category == 'spam'].Message)
ham = list(df.loc[df.Category == 'ham'].Message)

# print(spam[0])
# print(ham[0])

# Создаем список 0 элементом которого будут все спамные сообщения. А 1 и пооследующие элементы - 
# это обычные сообщения по отдельности (каждый элемент списка начиная с 1 это отдельно НЕ спамное сообщение)
messages = []
messages = ham
messages.insert(0, ' '.join(spam))
# message[0]
```

```python
# Приведем сообщения списка messages к нижнему регистру
messages_lower_case = [message.lower() for message in messages]

# print(messages_lower_case[0])
```

```python
# Уберем все служебные символы из списка messages
messages_only_words = [re.sub('[\W_]+',' ', message) for message in messages_lower_case]

# print(messages_only_words[0])
```

```python
# Разделим сообщения из списка messages на слова
messages_splitted = [message.split() for message in messages_only_words]

# print(messages_splitted[0])
```

```python
# Удалим стоп-слова из списка messages
from nltk.corpus import stopwords

stopwords_set = set(stopwords.words('english'))

messages_without_sw = [[word for word in message if word not in stopwords_set] for message in messages_splitted]

# print(messages_without_sw[0])
```

```python
# Лемматизируем слова в списке messages
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()

messages_lemmatized = [[wordnet_lemmatizer.lemmatize(word) for word in message] for message in messages_without_sw]

# print(messages_lemmatized[0])
```
