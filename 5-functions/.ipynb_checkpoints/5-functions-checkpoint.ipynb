№: 10006, тип: insurance, владелец: Аристарх Павлов, полка хранения: 2

```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}

# Задание 1. Пункт 1.
def find_user_by_numdoc (numdoc):
    """
    Функция принимает на вход и помещает в переменную numdoc номер документа пользователя.
    Затем перебирает список documents и сравнивает значение переменной numdoc со значением ключа number
    из текущего элемента списка documents. 
    Если найдено совпадение, переменной users присваивается значением ключа name
    из текущего элемента списка documents. 
    Если совпадение не найдено, переменная user_by_numdoc сохраняет присвоенное ей начальное значение
    "Пользователь не найден". По окончании перебора списка documents функция возвращает значение переменной user_by_numdoc.
    """
    user_by_numdoc = 'Пользователь не найден'
    index = 0
    for item in documents:
        if numdoc == item['number']:
            user_by_numdoc = 'Пользователь: ' + (item['name'])
            index = documents.index(item) # Индекс элемента списка documents, в котором найден искомы номер документа.
                                        # Индекс используется для удаления словаря из списка documents в функции dell_document
    return(user_by_numdoc, index)

# Задание 1. Пункт 2.
def find_shelf_by_numdoc (numdoc):
    """
    Функция принимает на вход и помещает в переменную numdoc номер документа пользователя.
    Затем происходит вызов функции find_user_by_numdoc, чтоб проверить существует ли пользователь с указанным номером документа.
    Если пользователь существует выполняется перебор словаря directories по ключам. Для каждого ключа происходит перебор
    его значений (т.е. перебор списка, который является значение для текущего ключа) и сравнение со значением переменной numdoc.
    Если совпадение найдено, функция прсваивает переменной shelf_by_numdoc текущий ключ.
    Если совпадение не найдено, переменная shelf_by_numdoc сохраняет присвоенное ей начальное значение
    "Пользователь не найден". По окончании перебора списка documents функция возвращает значение переменной shelf_by_numdoc.
    """
    shelf_by_numdoc = 'Пользователь не найден'
    index = '0'
    if find_user_by_numdoc (numdoc) != 'Пользователь не найден':
        for key in directories.keys():
            for item in directories.get(key):
                if numdoc == item:
                    shelf_by_numdoc = 'Номер полки: ' + key
                    index = key # Индекс - это ключ в словаре directories', среди элементов котрого найден искомы номер документа.
                                # Индекс используется для удаления по ключу значения словаря directories в функции dell_document
    return(shelf_by_numdoc, index)

# Задание 1. Пункт 3.
def show_full_information ():
    """
    Функция перебирает список documents (каждый элемент списка является словарем) и выводит значения ключей словаря.
    При этом сталкиваясь со значением number (номер пользователя), происходит вызов функции find_shelf_by_numdoc для
    поиска номера полки с данными о пользователе в словаре directories.
    """
    for item in documents:
        print('№: ' + item['number'] + ', Тип: ' + item['type'] + ', Владелец: ' + item['name'] + ', ' + 
              find_shelf_by_numdoc (item['number'])[0])

# Задание 1. Пункт 4.        
def add_new_shell(shell_number):
    """
    Функция принимает на вход номер полки и передает в переменную shell_number.
    Проверяет отсутсвие значения переменной shell_number среди ключей словаря directories.
    Если значения переменной shell_number среди ключей словаря directories нет,
    происходит добавление новой пары ключ:значение (ключ берется из переменной shell_number, значение - это пустой список).
    Затем выводится список ключей соваря directories.
    Если значения переменной shell_number среди ключей словаря directories уже есть, 
    то просто выводится список ключей соваря directories.
    """
    if shell_number not in directories.keys():
        directories[shell_number] = list()
        shell_number = str('Полка добавлена. Текущий перечень полок: ')
#         print('Полка добавлена. Текущий перечень полок: ', ', '.join(directories.keys()))
    else:
#         print('Такая полка уже существует.', ', '.join(directories.keys()))
        shell_number = str('Такая полка уже существует.')
    return(shell_number)

# Задание 1. Пункт 5.
def dell_shell(shell_number):
    """
    Функция принимает на вход номер полки и передает переменной shell_number.
    Затем происходит проверка наличия в словаре directories ключа, соответвующего заданному номеру полки.
    Если ключ есть, происходит проверка присутствия пустого значения по этому ключу (т.е. проверка пуста ли полка).
    Если значение у проверяемого ключа пустое, пара ключ-значение удаляется и выводится сообщение.
    """
    if directories.get(str(shell_number), 'none') == 'none':
#         print('Такой полки не существует. Текущий перечень полок: ', ', '.join(directories.keys()))
        shell_number = str('Такой полки не существует. Текущий перечень полок: ')
    elif len(directories.get(str(shell_number))) == 0:
        directories.pop(str(shell_number))
#         print('Полка удалена. Текущий перечень полок: ', ', '.join(directories.keys()))
        shell_number = str('Полка удалена. Текущий перечень полок: ')
    else:
#         print('На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: ', ', '.join(directories.keys()))
        shell_number = str('На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: ')
    return(shell_number)

# Задание 2. Пункт 1.
def add_document(doc_num, doc_type, doc_owner, shell_num):
    """
    Функция принимает на вход номер документа, тип документа, имя владельца, номер полки и передает значения переменным
    doc_num, doc_type, doc_owner, shell_num соответвенно. Затем проверяется существование поданногона вход номера полки 
    среди ключей слваря directories. Если полки нет, выводится соответсвующее сообщение. Если полка есть, происходит вызов функции
    find_user_by_numdoc, которая проверяет существует ли уже в базе пользователь с указанным номером документа. 
    Если такой пользователь не существует, он добавляется. Если пользователь ужу существует выводится сообщение об этом.
    """
    if shell_num not in directories.keys():
        result = str('Такой полки не существует. Добавьте полку командой as. Текущий список документов: ')
    elif find_user_by_numdoc (doc_num) == 'Пользователь не найден':
        documents.append({'type': doc_type, 'number': doc_num, 'name': doc_owner})
        directories[shell_num].append(doc_num)
        result = str('Документ добавлен. Текущий список документов: ')
    elif find_user_by_numdoc (doc_num) != 'Пользователь не найден':
        result = str('Владелец с указанным номером уже существует. Текущий список документов: ')
    return(result)

# Задание 2. Пункт 2.
def dell_document(num_doc):
    if find_user_by_numdoc(num_doc)[0] != 'Пользователь не найден':
        del(documents[find_user_by_numdoc(num_doc)[1]])
        for item in directories[find_shelf_by_numdoc(num_doc)[1]]:
            if item == num_doc:
                directories[find_shelf_by_numdoc(num_doc)[1]].remove(item)
        result = str('Документ удален. Текущий список документов: ')
    else:
        result = str('Документ не найден в базе. Текущий список документов: ')
        show_full_information()
    return(result)

# Задание 2. Пункт 3.
def mov_document (num_doc, shell_num):
    if find_user_by_numdoc(num_doc)[0] == 'Пользователь не найден':
        print('Документ не найден в базе. Текущий список документов: ')
        show_full_information()
    elif directories.get(str(shell_num), 'none') == 'none':
        print('Такой полки не существует. Текущий перечень полок: ', ', '.join(directories.keys()))
    else:
        directories[shell_num].append(num_doc)
        print('Документ перемещен. Текущий список документов: ')
        show_full_information()
        
# Функция запуска.
def work (command):
    if command == 'q':
        return False
    elif command == 'p':
        print(find_user_by_numdoc(input('Укажите номер документа: '))[0])
    elif command == 's':
        print(find_shelf_by_numdoc (input('Укажите номер документа: '))[0])
    elif command == '/':
        show_full_information()
    elif command == 'ads':
        print(add_new_shell(input('Укажите номер полки: '))  + ' ' + ', '.join(directories.keys()))
    elif command == 'ds':
        print(dell_shell(input('Укажите номер полки: '))  + ' ' + ', '.join(directories.keys()))
    elif command == 'ad':
        print(add_document(input('Введите номер документа: '), input('Введите тип документа: '), 
                           input('Введите владельца документа: '), input('Введите полку для хранения: ')))
        show_full_information ()
    elif command == 'd':
        print(dell_document(input('Укажите номер документа: ')))
        show_full_information()
    elif command == 'm':
        mov_document(input('Введите номер документа: '), input('Введите номер полки: '))
    else:
        print('Команда не поддерживается!')
    
```

```python
# Циклический запуск функции work. Работает пока не будет введена команда q
while True:
    if work(input('Введите команду: ').lower()) == False:
        break
```
