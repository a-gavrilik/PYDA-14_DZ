```python
# Были взяты для анализа следующие столбцы с данными


#   1:  surgery?
#           1 = Yes, it had surgery
#           2 = It was treated without surgery

#   2:  Age
#           1 = Adult horse
#           2 = Young (< 6 months)

#   3:  Hospital Number
#           - numeric id
#           - the case number assigned to the horse
#             (may not be unique if the horse is treated > 1 time)
            
#   5:  pulse
#           - linear
#           - the heart rate in beats per minute
#           - is a reflection of the heart condition: 30 -40 is normal for adults
#           - rare to have a lower than normal rate although athletic horses
#             may have a rate of 20-25
#           - animals with painful lesions or suffering from circulatory shock
#             may have an elevated heart rate
            
#   7:  temperature of extremities
#           - a subjective indication of peripheral circulation
#           - possible values:
#                1 = Normal
#                2 = Warm
#                3 = Cool
#                4 = Cold
#           - cool to cold extremities indicate possible shock
#           - hot extremities should correlate with an elevated rectal temp.

#  13: abdominal distension
#           - An IMPORTANT parameter.
#           - possible values
#                1 = none
#                2 = slight
#                3 = moderate
#                4 = severe
#           - an animal with abdominal distension is likely to be painful and
#             have reduced gut motility.
#           - a horse with severe abdominal distension is likely to require
#             surgery just tio relieve the pressure

#  18: abdomen
#           - possible values
#                1 = normal
#                2 = other
#                3 = firm feces in the large intestine
#                4 = distended small intestine
#                5 = distended large intestine
#           - 3 is probably an obstruction caused by a mechanical impaction
#             and is normally treated medically
#           - 4 and 5 indicate a surgical lesion

#  23: outcome
#           - what eventually happened to the horse?
#           - possible values:
#                1 = lived
#                2 = died
#                3 = was euthanized
```

```python
# Номер госпиталя
print(f'Мода: {horse.hospital_num.mode()[0]}')
len(horse.hospital_num.unique())

# Вывод: Существенного перевеса в сторону госпиталя с каким-то определенным номером нет.
# Почти одинаковое количество раз обращались во все госпитали. Причем самих госпиталей много (уникальных номеров 284).

# Возможно, это связано с тем, что существует какое-то центральное отделение, 
# которое равномерно распределяет лошадей с заболеваниями
# по госпиталям так, чтоб не создавать нагрузку в каком-то конкретном госпитале. Например, если существует хорошая ветеринарная клиника
# с развитой сетью точек госпитализации в регионе, где очень много лошадей.

# Если же статистика дана по госпиталям в рамках большой географической зоны (страна, регион, несколько стран), 
# то на основании имеющихся данных можно предположить, что в каждый госпиталь не часто обращаются с лошадьми, 
# которые страдают от кишечных коликов. Если в этой географической зоне много лошадей, 
# то это может быть признаком хорошего ухода за животными.
# Либо признаком того, что лошади не часто едят что попало (чистоплотные животные, избирательные в пище, посторонние и 
# не знающие люди их не кормят и т.д.)

# Отсутствие перевеса в сторону одного госпиталая может говорить о том, что все госпитали более-менее равнозначны.
# Если б был перевес в сторону госпиталя с каким-то определенным номером, то
# можно было бы предположить, что в него часто обращаются, поскольку он считается лучшим. 
# Либо лечение в данном госпитале наиболее оптимально по стоимости. Либо именно этот госпиталь специализируется на лошадях или
# на кишечных инфекциях. Либо в регионе, где этот госпиталь находится, многие занимаются лошадьми.
```

```python
# Пульс
print(f'Максимальный: {max(horse.pulse)}')
print(f'Минимальный: {min(horse.pulse)}')
print(f'Размах: {max(horse.pulse) - min(horse.pulse)}')
print(f'Ср.арифм.: {horse.pulse.mean()}')
print(f'Мода: {horse.pulse.mode()[0]}')
print(f'Медиана: {horse.pulse.median()}')
print(f'СКО: {horse.pulse.std()}')
print(f'Межквартильный размах: {horse.pulse.quantile(0.75) - horse.pulse.quantile(0.25)}')

#Вывод: Пульс довольно изменчив (СКО = 28.6), значения сильно размазаны относительно среднего. 
# Довольно большой размах. Мода выше стандартного номального показателя (30-40) пульса для взрослой лошади. 
# Также, как и медиана. Т.е. вцелом пульс у лошадей, которые попали в анализ завышенный. 
# Поскольку рассматриваются животные с кишечными коликами, большинству из которых была сделана операция,
# можно предположить, что увеличение пульса коррелирует с налилием и развитием заболеваний желудка. 
# Получается желудочные колики создают повышенную нагрузку на сердце. Хотя, это, вроде бы очевидно - 
# при постоянном плохом самочувствии нагрузка на сердце увеличивается и пульс должен подниматься. 
# Если пульс упал - это скорее признак, что все совсем плохо.
```

```python
# Температура конечностей

print(f'Мода: {horse.temp_of_extremities.mode()[0]}')
print(f'Медиана: {horse.temp_of_extremities.median()}')
horse.temp_of_extremities.value_counts()

# Вывод: у многих живоьных сильно повышеная температура конечностей (значение 3), что является признаком повышенной 
# ректальной температуры. А повышенная ректальная температура, вероятно, признак заболевания. Имею ввиду, что проблема 
# со здоровьем у животного не только что проявилая. А, раз поднялась высокая температура, заболевание уже подзатянулось.
# На раннем этапе болезни владельцы этих лошадей либо не заметили проблему. Либо заметили изменение в поведении 
# животного, но не придали этому значения. Либо решили, что сами справятся или само пройдет.


# У большинства лошадей не наблюдалось холодных конечностей, что являлось бы признаком шока (скорее всего болевого шока,
# поскольку при коликах животному может быть больно). Т.е. у большинство животных болевого шока не было.
```

```python
# Растяжение брюшной полости

print(f'Мода: {horse.abdominal_distension.mode()[0]}')
print(f'Медиана: {horse.abdominal_distension.median()}')
horse.abdominal_distension.value_counts()

# Вывод: у большинства животных растяжение брюшной полости не наблюдается либо оно умеренно-слабое. 
# Т.е. не тяжелая форма, при которой требуется операция. Во-первых, раз есть такая метрика, то можно предположить, 
# что одна из причин появления колик - это растяжение брюшной полости и малая подвижность кишечника (так и в пояснении написано).
# При этом лошадей часто оперируют, но вероятно не из-за колик, которые появились по причине растяжения брюшной полости.
# Причина появления колик в чем-то другом - например, питание.
```

```python
# Брюшная полость

print(f'Мода: {horse.abdomen.mode()[0]}')
print(f'Медиана: {horse.abdomen.median()}')
horse.abdomen.value_counts()

# Вывод: довольно часто встречаются значения 4 и 5. Что свидельствует о растянутом толстом и тонком кишечнике - чаще толстом (5).
# В пояснениях указано, что это результат хирургического поражения. Насколько понимаю, имеется ввиду перенесенная ранее операция.
# Могу предположить, что у многих лошадей ранее были проблемы, которые решались хирургическим путем. 
# Возможно, это хирургическое вмешательство было не достаточно качественным и проффесиональным 
# Как вариант, владельцы животных решали проблему "сами и своими силами". Т.е. текущие проблемы (кишечные колики) могут быть 
# связаны с перенесенными ранее заболеваниями и не правильным лечененим.
```

```python
# Итоги

print(f'Мода: {horse.outcome.mode()[0]}')
print(f'Медиана: {horse.outcome.median()}')
horse.outcome.value_counts()

# Вывод: большинство лошадей остались живы. Значит колики не то заболевание, которое часто приводит к тому, что животное погибает.
# Также это признак того, что чаще владельцы все таки успевают обратиться к ветеринару.
```

```python
# Задание 2 - Выбросы

q1 = horse.pulse.quantile(0.25)
q3 = horse.pulse.quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = horse[horse.pulse.between(lower_bound, upper_bound, inclusive=True)].sort_values('pulse')

pd.concat([horse, remove_outliers]).drop_duplicates(keep=False).sort_values('pulse')

# Вывод: 
# 1) значения 150, 160, 164, 184 являются слишком высокими и потенциально могут считаться выбрасами. 
# На мой взгляд данные выбросы стоит оставить, поскольку данные значения действительно могли быть зафиксированы, 
# т.е. это могут реальные данные, не ошибки. Нормальный пульс лошади в состоянии покоя 25-44. При работе в районе 70.
# При интенсивной нагрузке (когда лошади что-то угрожает или во время соревнований, т.е. когда нужно бежать из зо всех сил) 
# пульс у лошадей может подниматься до 220-225.

# При этом в таблице приведены данные лошай с заболеванием. 
# Данное заболевание может быть очень болезненным - лошади очень больно, она испугана и пульс растет.
# Так же в клинике животному тоже может быть страшно и пульс тоже растет.
# Т.е. выбросы могут быть реальными значениями (не ошибки) при определенных обстоятельствах.

# 2) Есть еще выбросы по возрасту. В столбец возраст записывается 1 или 2. Но в некоторых строках возраст равен 9.
# Это явно ошибка. Возможно возраст лошади 9 лет или 9 месяцев, скорее 9 лет (лошади живут по 25-30 лет).
# В любом случае - это по классификации в таблице взрослое животное. Поэтому предлагаю 9 заменить на 1 (1 - это взрослые),
# чтоб не терять данные.
```

```python
# Меняем возраст 9 на 1
horse.age.replace(9, 1, inplace=True)
```

```python
# По значению пульса пропуски есть. Предполагаю, что пульс зависит от тяжести состояния лошади. Тяжесть состояния определяет
# потребуется ли операция. Поэтому предложил бы заполнить медианой в зависимости от того потребовалась ли операция.
# Медианой, поскольку показатель пульса числовой. Также есть выбросы по пульсу и они значимые (высокие).
# Поэтому, чтоб не исказить статистику предлагаю заполнить медианой с группировкой по столбцу surgery.

horse[horse['pulse'].isna()]
```

```python
horse.temp_of_extremities.fillna(horse.temp_of_extremities.mode()[0], inplace=True)
```

```python
# Предложил бы заполнить модой (поскольку показатель категорийный ) в зависимости от показателя по операции (поскольку в пояснении написано, 
# что лошади с сильным растяжением живота, вероятно, потребуется операция.

horse[horse['abdominal_distension'].isna()]
```

```python
horse.abdominal_distension.fillna(horse.groupby('surgery').abdominal_distension.transform('median'), inplace=True)
```

```python
# Нагуглил, что растяжение толстого и тонкого кишечника (в первом задания был сделан вывод, что именно эта проблема чаще 
# всего встречается в показателях) - это довольно серьезный недуг и может привести к смерти животного. 
# Поэтому предложил бы заполнить пропуски модой (поскольку показатель категорийный) в зависимости от значения столбца outcome 
# (итог: 1-выжил, 2-умер, 3-эфтаназия). Т.е. с группировкой по столбцу outcome.

horse[horse['abdomen'].isna()]
```
