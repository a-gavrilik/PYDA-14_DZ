# Задание 1

Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
- оценка 2 и меньше - низкий рейтинг
- оценка 4 и меньше - средний рейтинг
- оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class

```python
import pandas as pd
```

```python
path = '/Users/a.gavrilik/Desktop/Data Analytics/PYDA-14_DZ/12-biblioteka_pandas/ml-latest-small/'

def classified (row_rating):
    """Классифицмруем ретинг фильма согласно условию:
    2 и меньше - Low
    4 и меньше - Midle
    4.5 и 5 - Hight"""
    if row_rating <= 2:
        return 'Low'
    elif row_rating <= 4:
        return 'Midle'
    elif row_rating > 4:
        return 'Hight'

# Объединенная таблица rating_5, которая включает ratings.csv + movies.csv
movie_rat = pd.read_csv(path + 'ratings.csv').merge(pd.read_csv(path + 'movies.csv'), on = 'movieId')

# Применяем функцию classified к каждому значению столбца raiting в movie_rat
movie_rat['class'] = movie_rat['rating'].apply(classified)
movie_rat.head()
```

# Задание 3 (бонусное)

Есть мнение, что “раньше снимали настоящее кино, не то что сейчас”. Ваша задача проверить это утверждение, используя файлы с рейтингами фильмов из прошлого домашнего занятия (файл ratings.csv из базы https://grouplens.org/datasets/movielens). Т. е. проверить верно ли, что с ростом года выпуска фильма его средний рейтинг становится ниже.

При этом мы не будем затрагивать субьективные факторы выставления этих рейтингов, а пройдемся по следующему алгоритму:

1. В переменную years запишите список из всех годов с 1950 по 2010.

2. Напишите функцию production_year, которая каждой строке из названия фильма выставляет год выпуска. Не все названия фильмов содержат год выпуска в одинаковом формате, поэтому используйте следующий алгоритм:

    1. для каждой строки пройдите по всем годам списка years
    2. если номер года присутствует в названии фильма, то функция возвращает этот год как год выпуска
    3. если ни один из номеров года списка years не встретился в названии фильма, то возвращается 1900 год

3. Запишите год выпуска фильма по алгоритму пункта 2 в новый столбец ‘year’

4. Посчитайте средний рейтинг всех фильмов для каждого значения столбца ‘year’ и отсортируйте результат по убыванию рейтинга

```python
path = '/Users/a.gavrilik/Desktop/Data Analytics/PYDA-14_DZ/12-biblioteka_pandas/ml-latest-small/'

# Объединенная таблица rating_5, которая включает ratings.csv + movies.csv
movies_rating = pd.read_csv(path + 'ratings.csv').merge(pd.read_csv(path + 'movies.csv'), on = 'movieId')

# Перебираем значения слобца title из movies_rating
# Разбиваем каждое значение слобца title на слова, получаем список слов. 
# Берем последнее слово списка (т.е. подразумеваем, что это цифру года без скобок).
# Проверяем является ли последнее слово без скорол цифрой и есть ли в списке years такой год. 
# Если это цифра и ее нет в years - добавляем год в список years.
years = list()
for value in movies_rating['title']: 
    if (value.strip().split(' ')[-1][1:5].isdigit()) and (value.strip().split(' ')[-1][1:5] not in years): 
        years.append(value.strip().split(' ')[-1][1:5])
        
print(years)

# Получение списка years с годами написал, но не использовал для решения задачи.
# Получилось обойтись без него.
```

```python
# Функция получает на вход строку. Отбраывает пробелы в начале и конце строки. 
# Разбивает строку на слова по пробелу. Формирует из слов списк.
# Берет последний элемент списка без первого и последнего символа (возможно - это будет цифра года).
# Проверяет является ли полученное значение числом через isdigit().
# Если это число - возвращает его. Если не число - возвращает 1900
def production_year(row_title):
    if row_title.strip().split(' ')[-1][1:5].isdigit():
        return row_title.strip().split(' ')[-1][1:5]
    else:
        return '1900'
    
movies_rating['year'] = movies_rating['title'].apply(production_year)
movies_rating.head()
```

```python
# Функция выполняет группировку рейтингов по годам и вычисляет средний рейтинг для каждого года.
def groupby_and_mean(row_year):
    return row_year.rating.mean()

# sort_values(ascending = False) - сортировка полученного Series по убыванию
movies_rating.groupby('year').apply(groupby_and_mean).sort_values(ascending = False).reset_index()
```
