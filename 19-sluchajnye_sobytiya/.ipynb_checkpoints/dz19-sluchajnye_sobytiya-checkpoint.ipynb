# Дополнительное задание повышенной сложности. 

Теоретически рассчитать средний выигрыш (проигрыш) и сравнить с результатами моделирования

```python
import numpy as np
np.random.seed(42)

money = 0 # Сумма выигрыша/проигрыша
trials = 1000000 # Кол-во повторений (сколька раз сыграл игрок)
success = 0 # Кол-во успехов (сколько раз игрок хоть что-нибудь выиграл)
p = 8/1000 # Теоритическая вероятность успеха. Вероятность, что игрок хоть что-нибудь выиграл, т.е. выиграл хотя бы 2 рубля

for i in range(trials):
    money -= 1
    number = np.random.randint(1000)
    if number == 777:
        success += 1
        money += 200
    elif number == 999:
        success += 1
        money += 100
    elif number == 555:
        success += 1
        money += 50
    elif number == 333:
        success += 1
        money += 15
    elif number == 111:
        success += 1
        money += 10
    elif number in range(177, 1000, 100):
        success += 1
        money += 5
    elif number in range(107, 1000, 10):
        success += 1
        money += 3
    elif number in range(100, 1000, 100):
        success += 1
        money += 2
    elif number in range(110, 1000, 10):
#         success += 1 выигрыш в 1 рубль - это фактически не выигрыш, а игра в 0. Поэтому здесь успех не считаем.
        money += 1    

p_monte_carlo = success / trials    # Вероятность смоделированного успеха (частота по Монте Карло)    

print(money) # Сумма выигрыша/проигрыша

print(p, p_monte_carlo) # Теоритическая вероятность успеха, Вероятность смоделированного успеха

print(money/trials) # Сколько в среднем приобретает или теряет игрок за одну игру
```

Вывод: 
- для игрока игра в целом не выгода. Чем больше игрок играем тем большую сумму он проиграет. Например, даже если с первого раза получится выиграть 200 рублей (хоть это очень маловероятно), потратив 1 рубль. То при второй, третьей и т.д. попытках сумма проигрыша будет увеличваться.
- в среднем игрок теряет около 0,24-0,25 рубля при приблизительно 10000 и выше повторений игры. Если количество повторений игры меньше чем приблизительно 10000 раз, то игрок теряет в среднем больше.
